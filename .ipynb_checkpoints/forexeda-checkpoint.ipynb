# Add Target for Classification (Up / Down)
target of i day is the np.sign((open i+d) - (open i))

```python
df.head()
```

```python
df['actual_up_down'] = np.sign(df['open']-df['open'].shift(1))
df['log_returns'] = np.log(df['open'] / df['open'].shift(1))
df['ari_returns'] = (df['open'] / df['open'].shift(1)) - 1
df['target_future_1_direction'] = np.sign(df['open'].shift(-1) - df['open'])
```

```python
df.head(30)
```

# Add Features pandas

```python
df['20d_ma'] = df['open'].rolling(window=20).mean()
df['50d_ma'] = df['open'].rolling(window=50).mean()
df['Bol_upper'] = df['open'].rolling(window=20).mean() + 2* df['open'].rolling(window=20, min_periods=20).std()
df['Bol_lower'] = df['open'].rolling(window=20).mean() - 2* df['open'].rolling(window=20, min_periods=20).std()
df['Bol_BW'] = ((df['Bol_upper'] - df['Bol_lower'])/df['20d_ma'])*100
df['Bol_BW_200MA'] = df['Bol_BW'].rolling(window=50).mean()
df['20d_exma'] = df['open'].ewm(span=20).mean()
df['50d_exma'] = df['open'].ewm(span=50).mean()
```

```python
from talib import MA_Type
upper, middle, lower = talib.BBANDS(df['open'].values, matype=MA_Type.T3)
```

```python
df['upper'] = upper
df['middle'] = middle
df['lower'] = lower
df[datetime(2015,1,1):].plot(y=['open','upper', 'middle', 'lower'], figsize=(25,10))
```

```python
cols = []
for momentum in [15, 30, 60, 120]:
    col = 'average_log_return_sign_{}'.format(momentum)
    df[col] = np.sign(df['log_returns'].rolling(momentum).mean()).shift(1) #the sign of the average returns of the last x candles
    cols.append(col)
```

```python
df.head(20)
```

```python
strats = ['log_returns']
for col in cols:
    strat = 'strategy_{}'.format(col)
    df[strat] = df[col] * df['log_returns'] #shift last sign one and multiply by return
    strats.append(strat)
```

```python
strats
```

```python
fig, axes = plt.subplots()
cum_returns = df[strats].dropna().cumsum().apply(np.exp)-1 #you can add log returns and then transpose back with np.exp
cum_returns.plot(ax=axes)
plt.legend(loc='best')
```
